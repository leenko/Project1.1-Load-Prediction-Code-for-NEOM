```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
def parser(x):
    '''Date Parser'''
    return pd.to_datetime(x)
```

```python
train = pd.read_csv('train.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
train.index = pd.DatetimeIndex(train.index).to_period('h')
train.head(10)
```

```python
test = pd.read_csv('test.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
test.index = pd.DatetimeIndex(test.index).to_period('h')
test.head(10)
```

```python
ax=train.plot(title='Electricity Load vs Date ')
ax.set_xlabel("Date")
ax.set_ylabel('Electricity Load')
plt.show()
```

```python
autocorrelation_plot(train)
```

```python
# fit model
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
# plot residual errors
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
#Forecast with 10 step hopping(Not Very Much Accurate). This will take less time but accuracy is low
history2 = [x for x in train]
forecast = list()
out_arr2 = []
step_c = 10

for t in range(0, 100, step_c):
    model = ARIMA(history2, order=(5,1,0))
    model_fit = model.fit(disp=0)
    yhat = model_fit.forecast(steps=step_c, alpha=0.99)[0]
    for i in range(step_c):
        forecast.append(yhat[i])
        obs = test[t + i]
        out_arr2.append([obs, yhat[i]])
        history2.append(obs)
        print('predicted=%f, expected=%f' % (yhat[i], obs))
```

```python
error = mean_squared_error(test[:len(forecast)], forecast)
print('Test MSE: %.3f' % error)
```

```python
# plot for Forecast With Hop
samples = 100
exp = test[:samples]
exp.index = exp.index.to_timestamp().to_pydatetime()
plt.plot(exp)
plt.plot(exp.index, forecast[:samples], color='red')
plt.show()
```

```python
out_df = pd.DataFrame(out_arr2, index=test.index[:len(forecast)], columns=['Expected', 'Predicted'])
d1=out_df
out_df.head(10)
```

```python
predicted_=np.array(d1['Predicted']).astype(float)
expected_=np.array(d1["Expected"]).astype(float)

error=np.mean(np.abs((expected_ - predicted_) /expected_)) 
print("Train Error : ",error)
```

```python
def daterange(start_date, end_date):
    delta = timedelta(hours=1)
    while start_date < end_date:
        yield start_date
        start_date += delta
        
from datetime import datetime, timedelta
l=[]

start_date = datetime(2017, 1, 1, 1, 00)
end_date = datetime(2022, 1, 1, 00, 00)
for single_date in daterange(start_date, end_date):
    l.append(single_date.strftime("%m/%d/%Y %H:%M"))
    
out_date=pd.DataFrame(np.array(l),columns=['date'])
```
